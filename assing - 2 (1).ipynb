```python
import numpy as np
from scipy import stats
```

```python
#The data which is provied i have taken that 
data= [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 
        1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
# and then find its total mean ,std,len , and (n:no of samples)
n = len(data)
mean = np.mean(data)
std_dev = np.std(data,ddof=1)
```

```python
#here they told to fing 99% of confidence interval 
#we already kmow the (t-distibution) so here .I taken percentile point funbction 
confidence = 0.99
alpha = 1 - confidence
df = n - 1  # degrees of freedom
t_critical = stats.t.ppf(1 - alpha/2, df)
```

```python
#the margin of error
margin_of_error_t = t_critical * (std_dev / np.sqrt(n))
ci_lower_t = mean - margin_of_error_t
ci_upper_t = mean + margin_of_error_t
```

```python
print(f"99% CI using sample std dev (t-distribution): ({ci_lower_t:.4f}, {ci_upper_t:.4f})")
```

```python
#same here they given population erroe (0.2) so (z distribution ) after that fin
pop_std_dev = 0.2
z_critical = stats.norm.ppf(1 - alpha/2)

margin_of_error_z = z_critical * (pop_std_dev / np.sqrt(n))
ci_lower_z = mean - margin_of_error_z
ci_upper_z = mean + margin_of_error_z
```

```python
print(f"99% CI using sample known population std dev (z-distribution): ({ci_lower_t:.4f}, {ci_upper_t:.4f})")
```
